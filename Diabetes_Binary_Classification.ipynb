```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/diabetes (3).csv")
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
#Setting target variable and predictor variable
X = data.drop('Outcome',axis=1)
y = data['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=np.random.seed(10))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from keras import layers
from keras import models
model = models.Sequential()
model.add(layers.Dense(16, activation = 'relu', input_shape = (2,))) #Right now these neuron values are taken radnomly
model.add(layers.Dense(16, activation = 'relu'))#HIdden layer
model.add(layers.Dense(1, activation = 'sigmoid')) #Output is binary so one neuron, with sigmoid AF
```

```python
model.summary() #Shows the parameters and the shapes (Params = weights and biases)
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(loss='binary_crossentropy',   #beacuse we are doing binary prediction
             metrics=['accuracy'])
```

```python
accuracy = model.fit(X_train, y_train, epochs = 10)
```

```python
plt.plot(accuracy.history['loss'])
```

```python
print(accuracy.history.keys())
```

```python
#Changing the epochs
accuracy = model.fit(X_train, y_train, epochs = 25)
plt.plot(accuracy.history['loss'])
plt.title("Loss")
```

```python
#Changing the neurons
from keras import layers
from keras import models
model = models.Sequential()
model.add(layers.Dense(16, activation = 'relu', input_shape = (2,))) #Right now these neuron values are taken radnomly
model.add(layers.Dense(16, activation = 'relu'))#HIdden layer
model.add(layers.Dense(16, activation = 'relu'))#HIdden layer

model.add(layers.Dense(16, activation = 'relu'))#HIdden layer

model.add(layers.Dense(16, activation = 'relu'))#HIdden layer

model.add(layers.Dense(1, activation = 'sigmoid')) #Output is binary so one neuron, with sigmoid AF


from tensorflow.keras.optimizers import Adam
model.compile(loss='binary_crossentropy',   #beacuse we are doing binary prediction
             metrics=['accuracy'])


accuracy = model.fit(X_train, y_train, epochs = 1000)
plt.plot(accuracy.history['loss'])
plt.title("Loss")
```

```python
#Changing the activation function to LeakyRelu for the hidden layers

#Changing the neurons
from keras import layers
from keras import models
model = models.Sequential()
model.add(layers.Dense(16, activation = 'relu', input_shape = (2,))) #Right now these neuron values are taken radnomly
model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer
model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer

model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer

model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer

model.add(layers.Dense(1, activation = 'sigmoid')) #Output is binary so one neuron, with sigmoid AF


from tensorflow.keras.optimizers import Adam
model.compile(loss='binary_crossentropy',   #beacuse we are doing binary prediction
             metrics=['accuracy'])


accuracy = model.fit(X_train, y_train, epochs = 1000)
plt.plot(accuracy.history['loss'])
plt.title("Loss")


#The accuracy has improved 
```

```python
#Changing the activation function to LeakyRelu for the hidden layers

#Changing the neurons
from keras import layers
from keras import models
model = models.Sequential()
model.add(layers.Dense(16, activation = 'relu', input_shape = (2,))) #Right now these neuron values are taken radnomly
model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer
model.add(layers.Dense(16, activation = 'leaky_relu'))#HIdden layer


model.add(layers.Dense(1, activation = 'sigmoid')) #Output is binary so one neuron, with sigmoid AF


from tensorflow.keras.optimizers import Adam
model.compile(loss='binary_crossentropy',   #beacuse we are doing binary prediction
             metrics=['accuracy'])


accuracy = model.fit(X_train, y_train, epochs = 100)
plt.plot(accuracy.history['loss'])
plt.title("Loss")


#The accuracy has decreased further
```
